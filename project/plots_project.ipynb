```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
Ls = [10] # [10, 20, 30, 40]
Ni = 1000
Nbins = 20
Niter = Ni * Nbins
corr_type = 'all' # all, final

Rs = [np.arange(1, L) for L in Ls]
end = None

plt.figure(dpi=144)
for i, L in enumerate(Ls):
    data = np.load('data/L%s_N%s_%sx%sbins_%s.npz' % (L, Niter, Ni, Nbins, corr_type))
    corrs, errs = data['corrs'], data['errs']
    
    plt.errorbar(Rs[i][:end], corrs[0][:end], errs[0][:end] * np.sqrt(Ni),
                 fmt='.-', label='Horizontal')
    plt.errorbar(Rs[i][:end], corrs[1][:end], errs[1][:end] * np.sqrt(Ni),
                 fmt='.-', label='Vertical')

plt.title('Correlations in same direction', fontsize=14)
plt.xlabel('R', fontsize=10)
plt.ylabel('$<s_0 s_r>$', fontsize=10)
# plt.xscale('log')
# plt.yscale('log')
plt.legend()

plt.savefig("plots/L%s_N%s_%sx%sbins%s.png" % (Ls[0], Niter, Ni, Nbins, '_testsamedir'))
plt.show()
```

```python
Ls = [10] # [10, 20, 30, 40]
Ni = 1000
Nbins = 20
Niter = Ni * Nbins
corr_type = 'final' # all, final

Rs = [np.arange(1, L) for L in Ls]
Rsd = [np.sqrt(1/2) * np.arange(1, 2*L) for L in Ls] #diagonal
end = None

plt.figure(dpi=144)
for i, L in enumerate(Ls):
    data = np.load('data/L%s_N%s_%sx%sbins_%s.npz' % (L, Niter, Ni, Nbins, corr_type))
    corrs, errs = data['corrs'], data['errs']
    
    plt.errorbar(Rs[i][:end], np.abs(corrs[0][:end]), errs[0][:end] * np.sqrt(Ni),
                 fmt='.-', label='Same direction')
    plt.errorbar(Rs[i][:end], np.abs(corrs[1][:end]), errs[1][:end] * np.sqrt(Ni),
                 fmt='.-', label='Perpendicular')
    plt.errorbar(Rsd[i][1:end:2], np.abs(corrs[2][1:end:2]), errs[2][1:end:2] * np.sqrt(Ni),
                 fmt='.-', label='Diagonal (same dir)')
    plt.errorbar(Rsd[i][:end:2], np.abs(corrs[2][:end:2]), errs[2][:end:2] * np.sqrt(Ni),
                 fmt='.-', label='Diagonal (perp. dir)')

plt.title('%s x %s, #iter: %s' % (Ls[0], Ls[0], Niter), fontsize=14)
plt.xlabel('$\log(R)$', fontsize=10)
plt.ylabel('$\log \ |<s_0 s_r>|$', fontsize=10)
plt.xscale('log')
plt.yscale('log')
plt.legend()

plt.savefig("plots/L%s_N%s_%sx%sbins%s.png" % (Ls[0], Niter, Ni, Nbins, '_final'))
plt.show()
```
