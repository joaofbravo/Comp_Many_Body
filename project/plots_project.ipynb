```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
Ls = [10] # [10, 20, 30, 40]
Ni = 1000
Nbins = 20
Niter = Ni * Nbins
corr_type = 'all' # all, final

Rs = [np.arange(1, L) for L in Ls]
end = None

plt.figure(dpi=144)
for i, L in enumerate(Ls):
    data = np.load('data/L%s_N%s_%sx%sbins_%s.npz' % (L, Niter, Ni, Nbins, corr_type))
    corrs, errs = data['corrs'], data['errs']
    
    plt.errorbar(Rs[i][:end], corrs[0][:end], errs[0][:end],
                 fmt='.-', label='Horizontal')
    plt.errorbar(Rs[i][:end], corrs[1][:end], errs[1][:end],
                 fmt='.-', label='Vertical')

plt.title('Correlations along spin direction', fontsize=14)
plt.xlabel('r', fontsize=10)
plt.ylabel('$<s_0 s_r>$', fontsize=10)
# plt.xscale('log')
# plt.yscale('log')
plt.legend()

plt.savefig('plots/L%s_N%s_%sx%sbins%s.png' % (Ls[0], Niter, Ni, Nbins, '_testsamedir'))
plt.show()
```
