```python
import pickle
import generate_data
import matplotlib.pyplot as plt
```

```python
data = generate_data.load_data("data_ising_3D_largeL.pkl")
print(data[8].keys())
```

```python
def plot_obs(key, ax=None):
    if ax is None:
        plt.figure(figsize=(10, 7))
        ax = pl.gca()
    for L in data['Ls']:
        data_L = data[L]
        ax.errorbar(data_L['Ts'], data_L[key][:, 0], data_L[key][:, 1], label='L={L: 3d}'.format(L=L))
    ax.set_xlabel('T')
    ax.set_ylabel(key)
    ax.legend()
    #pl.show('web-agg')
    return ax
    
```

```python
plot_obs('E')
plot_obs('M')
plot_obs('absM')
plot_obs('C')
plot_obs('chi')
ax = plot_obs('UB')
#ax.set_xlim(2.2, 2.4)
pl.show()
```
